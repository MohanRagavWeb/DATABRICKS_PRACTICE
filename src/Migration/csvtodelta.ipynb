```python
df = spark.read.format("csv") \
    .option("header", True) \
    .option("inferSchema", True) \
    .load("/Volumes/workspace/default/my_volume_1/sales.csv")

display(df)
```

```python
df.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("sales_csv_to_delta")
```
