```python
df = spark.read \
.format("csv") \
.option("header",True) \
.option("inferSchema",True) \
.load("/Volumes/workspace/default/my_volume_2/governance.csv")


df.write \
.format("delta") \
.mode("overwrite") \
.saveAsTable("governance")
```

```python
%sql
SELECT * FROM governance;
```

Governance Technique 1 — Access Control

```python
%sql
GRANT SELECT
ON TABLE governance
TO `mohan.r@diggibyte.com`;
```

Governance Technique 2 — Column Level Security

```python
%sql
GRANT SELECT (id,name,email,city)
ON TABLE governance
TO `mohan.r@diggibyte.com`;
```

Governance Technique 3 — Row Level Security

```python
%sql
CREATE VIEW customers_chennai AS
SELECT *
FROM governance
WHERE city='Chennai';
```

```python
%sql
GRANT SELECT
ON VIEW customers_chennai
TO `mohan.r@diggibyte.com`;
```

```python
%sql
select * from customers_chennai
```

Revoke table access

```python
%sql
REVOKE SELECT
ON TABLE governance
FROM `mohan.r@diggibyte.com`;
```

```python
%sql
GRANT SELECT
ON VIEW customers_chennai
TO `mohan.r@diggibyte.com`;
```

Governance Technique 4 — Data Masking

```python
%sql
CREATE VIEW customers_masked AS
SELECT
id,
name,
concat('****@', split(email,'@')[1]) as email,
city,
salary
FROM governance;
```

```python
%sql
select * from customers_masked
```

Governance Technique 5 — Data Classification Tags

```python
%sql
ALTER TABLE governance
SET TAGS ('classification'='PII');
```
