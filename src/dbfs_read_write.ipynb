1️⃣ CSV

```python
df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .option("delimiter", ",") \
    .load("/path/file.csv")



df.write.format("csv") \
    .option("header", "true") \
    .mode("overwrite") \
    .save("/path/output/")
```

2️⃣ JSON

```python
df = spark.read.format("json") \
    .option("multiline", "true") \
    .load("/path/file.json")



df.write.format("json") \
    .mode("overwrite") \
    .save("/path/output/")
```

3️⃣ Parquet (Columnar format)

```python
df = spark.read.format("parquet") \
    .load("/path/file.parquet")


df.write.format("parquet") \
    .mode("overwrite") \
    .save("/path/output/")
```

4️⃣ Delta (Most common in Databricks)

```python
df = spark.read.format("delta") \
    .load("/path/delta_table")



df.write.format("delta") \
    .mode("overwrite") \
    .save("/path/delta_table")
```

5️⃣ Text File

```python
df = spark.read.format("text") \
    .load("/path/file.txt")


df.write.format("text") \
    .mode("overwrite") \
    .save("/path/output/")
```

6️⃣ ORC

```python
df = spark.read.format("orc") \
    .load("/path/file.orc")



df.write.format("orc") \
    .mode("overwrite") \
    .save("/path/output/")
```

Quick Comparison
| Format   | Type             | Used For                |
|----------|------------------|-------------------------|
| CSV      | Row-based        | Simple data exchange    |
| JSON     | Semi-structured  | APIs, logs              |
| Parquet  | Columnar         | Analytics               |
| Delta    | Lakehouse format | Production pipelines    |
| Text     | Raw text         | Logs                    |
| ORC      | Columnar         | Big data analytics      |

**Most Used Formats in Databricks**

In real projects:

CSV → Raw ingestion
JSON → Streaming / logs
Parquet → Analytics storage
Delta → Production tables
