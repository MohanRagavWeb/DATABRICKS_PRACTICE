# Sales Analysis Notebook

This notebook demonstrates basic Databricks operations.

Steps:
1. Create sample data
2. Perform transformations
3. Display results

```python
data = [
    (1,"Laptop",50000,2),
    (2,"Phone",20000,3),
    (3,"Tablet",15000,1),
    (4,"Headphones",3000,4)
]

columns = ["id","product","price","quantity"]

df = spark.createDataFrame(data, columns)

display(df)
```

```python
from pyspark.sql.functions import col

df = df.withColumn("total_amount", col("price") * col("quantity"))

display(df)
```

```python
df.groupBy("product").sum("total_amount").show()
```
