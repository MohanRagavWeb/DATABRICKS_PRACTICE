Silver Layer (Data Cleaning)

```python
from pyspark.sql.functions import col

# read bronze data
df = spark.read.format("delta") \
.load("/Volumes/workspace/default/my_volume_1/medallion/bronze")

display(df)

# cleaning step
df_clean = df.filter(col("price") > 0)

display(df_clean)

# write cleaned data
df_clean.write.format("delta") \
.mode("overwrite") \
.save("/Volumes/workspace/default/my_volume_1/medallion/silver")
```
