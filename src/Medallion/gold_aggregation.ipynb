Gold Layer (Business Aggregation)

```python
from pyspark.sql.functions import sum

# read silver data
df = spark.read.format("delta") \
.load("/Volumes/workspace/default/my_volume_1/medallion/silver")

display(df)

# business aggregation
df_gold = df.groupBy("product") \
.agg(sum("price").alias("total_sales"))

display(df_gold)

# save gold layer
df_gold.write.format("delta") \
.mode("overwrite") \
.save("/Volumes/workspace/default/my_volume_1/medallion/gold")
```
