Creating a Delta Table

```python
data = [(1,"Laptop",50000),(2,"Phone",20000)]

columns = ["id","product","price"]

df = spark.createDataFrame(data, columns)

df.write.format("delta") \
.mode("overwrite") \
.save("/Volumes/workspace/default/my_volume_1/delta_sales")
```

Reading Delta Table

```python
df = spark.read.format("delta") \
.load("/Volumes/workspace/default/my_volume_1/delta_sales")

display(df)
```
