```python
data = [
    (1,"Laptop","Chennai",60000),
    (2,"Mobile","Bangalore",20000),
    (3,"Tablet","Chennai",15000),
    (4,"Laptop","Delhi",65000)
]

columns = ["order_id","product","city","amount"]

df = spark.createDataFrame(data, columns)
```

```python
df.write.format("delta") \
.partitionBy("city") \
.saveAsTable("sales_partitioned")
```

```python
%sql
SELECT * FROM sales_partitioned
WHERE city = 'Chennai';
```

```python
%sql
SHOW PARTITIONS sales_partitioned;
```
