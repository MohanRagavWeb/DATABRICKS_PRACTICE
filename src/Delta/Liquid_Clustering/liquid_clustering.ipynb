```python
data = [
    (1, "C101", "Laptop", 60000, "Chennai"),
    (2, "C102", "Mobile", 20000, "Bangalore"),
    (3, "C101", "Headphones", 3000, "Chennai"),
    (4, "C103", "Laptop", 65000, "Hyderabad"),
    (5, "C104", "Tablet", 15000, "Delhi"),
    (6, "C102", "Mobile", 22000, "Bangalore"),
    (7, "C101", "Laptop", 61000, "Chennai")
]

columns = ["order_id","customer_id","product","amount","city"]

df = spark.createDataFrame(data, columns)

display(df)
```

```python
df.write.format("delta") \
.mode("overwrite") \
.saveAsTable("sales_liquid_demo")
```

```python
%sql
SELECT * FROM sales_liquid_demo;
```

Enable Liquid Clustering

```python
%sql
ALTER TABLE sales_liquid_demo
CLUSTER BY (customer_id);
```

```python
%sql
INSERT INTO sales_liquid_demo VALUES
(8,'C101','Mouse',1000,'Chennai'),
(9,'C102','Keyboard',2000,'Bangalore'),
(10,'C105','Laptop',70000,'Mumbai');
```

Run OPTIMIZE (Apply Clustering) This step physically reorganizes the files.

```python
%sql
OPTIMIZE sales_liquid_demo;
```
