1️⃣ Change Data Feed (CDF)

Change Data Feed tracks all row-level changes in a Delta table such as:

INSERT 
UPDATE
DELETE

It allows systems to read only changed data instead of the entire table.

```python
%sql
CREATE TABLE customers_feature
(
id INT,
name STRING,
city STRING
)
USING DELTA
TBLPROPERTIES (delta.enableChangeDataFeed = true);
```

```python
%sql
INSERT INTO customers_feature VALUES (1,'Ravi','Chennai'), (2,'Rajesh','Mumbai');
UPDATE customers_feature SET city='Bangalore' WHERE id=1;
DELETE FROM customers_feature WHERE id=1;
```

```python
%sql
SELECT * 
FROM table_changes('customers_feature', 0);
```

3️⃣ Generated Columns

```python
%sql
CREATE TABLE sales_generated_column
(
order_id INT,
price INT,
quantity INT,
total_price INT GENERATED ALWAYS AS (price * quantity)
)
USING DELTA;
```

```python
%sql
INSERT INTO sales_generated_column (order_id,price,quantity)
VALUES (1,500,2),(2,1000,3),(3,1500,4);
```

```python
%sql
select * from sales_generated_column
```

4️⃣ Row Tracking

```python
%sql
ALTER TABLE sales_generated_column
SET TBLPROPERTIES (delta.enableRowTracking = true);
```

```python
%sql
select * from sales_generated_column
```

| Feature            | Purpose                        |

| Row Tracking       | Track row-level history        |
| Type Widening      | Expand column data types       |
