```python
data=[(1,"laptop",1000),(2,"phone",500)]
columns=["id","product","price"]
df=spark.createDataFrame(data,columns)
display(df)
df.write.format("delta").mode("overwrite").save("/Volumes/workspace/default/my_volume_1/delta_sales_SE")
```

```python
new_data = [(3,"Tablet",15000,"Electronics")]

columns2 = ["id","product","price","category"]

df2 = spark.createDataFrame(new_data, columns2)

df2.write.format("delta") \
.mode("append")\
.option("mergeSchema", "true") \
.save("/Volumes/workspace/default/my_volume_1/delta_sales_SE")
```
