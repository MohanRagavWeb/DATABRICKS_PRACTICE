Step 1 — Create Delta Table

```python
data = [(1,"Laptop",50000),(2,"Phone",20000)]

columns = ["id","product","price"]

df = spark.createDataFrame(data, columns)

df.write.format("delta") \
.mode("overwrite") \
.save("/Volumes/workspace/default/my_volume_1/delta_sales_TT")
```

Step 2 — Append More Data

```python
data2 = [(3,"Tablet",15000)]

df2 = spark.createDataFrame(data2, columns)

df2.write.format("delta") \
.mode("append") \
.save("/Volumes/workspace/default/my_volume_1/delta_sales_TT")
```

```python
display(spark.read.format("delta").load("/Volumes/workspace/default/my_volume_1/delta_sales_TT"))
```

Step 3 — Read Old Version

```python
df_old = spark.read.format("delta") \
.option("versionAsOf", 0) \
.load("/Volumes/workspace/default/my_volume_1/delta_sales_TT")

display(df_old)
```

Viewing Delta History

```python
%sql
DESCRIBE HISTORY delta.`/Volumes/workspace/default/my_volume_1/delta_sales_TT`
```

```python
display(spark.read.format("delta").load("/Volumes/workspace/default/my_volume_1/delta_sales_TT"))
```

Restoring an Old Version

```python
%sql
RESTORE TABLE delta.`/Volumes/workspace/default/my_volume_1/delta_sales_TT`
TO VERSION AS OF 0
```
