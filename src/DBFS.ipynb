Visual Comparison

| Path           | Used For           | Storage Type   | Modern?   |
|----------------|--------------------|---------------|-----------|
| dbfs:/path     | Spark operations   | DBFS          | Legacy    |
| /dbfs/path     | Python file access | DBFS          | Legacy    |
| /Volumes/path  | Spark + Python     | Unity Catalog | ✅ Modern |

Creating a Mount Point (Old Method)

```python
dbutils.fs.mount(
  source = "abfss://sales@storageaccount.dfs.core.windows.net/",
  mount_point = "/mnt/sales-data",
  extra_configs = {
      "fs.azure.account.key.storageaccount.dfs.core.windows.net":
      "ACCESS_KEY"
  }
)
"""
After mounting:

dbfs:/mnt/sales-data
"""
```

Reading Data from Mount Point

```python
df = spark.read.csv(
    "dbfs:/mnt/sales-data/sales.csv",
    header=True,
    inferSchema=True
)

display(df)
```

Listing Mounted Storage

```python
dbutils.fs.mounts()
```

Unmounting Storage

```python
dbutils.fs.unmount("/mnt/sales-data")
```

## Mount Points vs Modern Volumes

| Feature         | Mount Points (Legacy) | Volumes (Modern)      |
|-----------------|----------------------|-----------------------|
| Storage Access  | DBFS                 | Unity Catalog         |
| Security        | Basic                | Fine-grained          |
| Governance      | Limited              | Advanced              |
| Recommended     | ❌ Legacy            | ✅ Modern             |
