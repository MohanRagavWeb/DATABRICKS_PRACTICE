```python
# Create multiselect widget
dbutils.widgets.multiselect(
    "products",
    "Laptop",
    ["Laptop", "Phone", "Tablet"],
    "Select Products"
)

# Get selected values
products = dbutils.widgets.get("products")

print("Selected products:", products)

# Convert string to list
product_list = products.split(",")

data = [
    ("Laptop",50000),
    ("Phone",20000),
    ("Tablet",15000),
    ("Laptop",52000),
    ("Phone",21000)
]

columns = ["product","price"]

df = spark.createDataFrame(data, columns)

df_filtered = df.filter(df.product.isin(product_list))

display(df_filtered)
```
