```python
# Create dropdown widget
dbutils.widgets.dropdown("env", "dev", ["dev", "test", "prod"], "Environment")

# Get selected value
env = dbutils.widgets.get("env")

print("Selected Environment:", env)

# Example logic based on environment
if env == "dev":
    print("Running in Development Environment")
elif env == "test":
    print("Running in Testing Environment")
else:
    print("Running in Production Environment")
```

| Position | Parameter | Meaning |
|----------|----------|--------|
| 1        | "env"    | Widget name (key) used in code |
| 2        | "dev"    | Default value shown when notebook starts |
| 3        | ["dev","test","prod"] | List of allowed options |
