```python
files=dbutils.fs.ls("/Workspace/Users/mohan.r@diggibyte.com/DATABRICKS_PRACTICE/src")

for file in files:
    print(file.path)
```

2️⃣ Copying Files

```python

dbutils.fs.cp(
"/Volumes/workspace/default/my_volume_1/students.csv",
"/Volumes/workspace/default/my_volume_2"
)

```

3️⃣ Moving Files

```python
dbutils.fs.mv("source_path","destination_path")
```

4️⃣ Deleting Files

```python
dbutils.fs.rm("path")

# To delete a folder recursively:
dbutils.fs.rm("path", True)
```

```python
dbutils.fs.mkdirs("/Workspace/Users/mohan.r@diggibyte.com/DATABRICKS_PRACTICE/src/dbutils_Commands/new_folder")
```

Summary

| Command              | Purpose           |
|----------------------|-------------------|
| dbutils.fs.ls()      | list files        |
| dbutils.fs.cp()      | copy files        |
| dbutils.fs.mv()      | move files        |
| dbutils.fs.rm()      | delete files      |
| dbutils.fs.mkdirs()  | create directory  |
