Create Widgets

```python
dbutils.widgets.text("file_name","sales.csv","Input File")
dbutils.widgets.text("run_date","2026-03-02","Run Date")
```

file_name → widget key

sales.csv → default value

File Name → label shown in UI

Read Widget Values

```python
file_name = dbutils.widgets.get("file_name")
run_date = dbutils.widgets.get("run_date")

print(file_name)
print(run_date)
```

Load Data from Volume

```python
df = spark.read.format("csv") \
.option("header","true") \
.option("inferSchema","true") \
.load(f"/Volumes/workspace/default/my_volume_1/{file_name}")

display(df)
```

Transform Data

```python
from pyspark.sql.functions import col

df_filtered = df.filter(col("date") == run_date)

display(df_filtered)
```

Write to Delta Table

```python
df_filtered.write.mode("overwrite") \
.format("delta") \
.saveAsTable("workspace.default.sales_filtered")
```

Verify Output

```python
df_out = spark.table("workspace.default.sales_filtered")

display(df_out)
```
